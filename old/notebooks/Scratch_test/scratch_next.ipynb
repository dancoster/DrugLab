```python
from scipy import stats
pvals = []
counts = med_lab.groupby(["ITEMID", "GENDER"]).count()["SUBJECT_ID"]
res = pd.DataFrame(counts)
res = res[res["SUBJECT_ID"]>100]
for i in res.index:
    med = i[0]
    gender = i[1]
    k_med = med_lab[med_lab["ITEMID"]==med]
    k_med = k_med[k_med["GENDER"]==gender]
    pvals.append(stats.mannwhitneyu(k_med["Before1"], k_med["After1"])[1])
res["MannwhitneyuTest"] = pvals
```

```python
med_lab_all = pd.read_csv(os.path.join(data, "mimiciii/1.4/preprocessed", "lab_med_all_stats_with_age_05112022.csv"))
```

```python
from scipy import stats
pvals = []
ttest = []
counts = med_lab_all.groupby(["ITEMID", "LAB_ITEMID"]).count()["SUBJECT_ID"]
res = pd.DataFrame(counts)
res = res[res["SUBJECT_ID"]>100]
for i in res.index:
    med = i[0]
    gender = i[1]
    k_med = med_lab_all[med_lab_all["ITEMID"]==med]
    k_med = k_med[k_med["GENDER"]==gender]
    pvals.append(stats.mannwhitneyu(k_med["Before1"], k_med["After1"])[1])
    ttest.append(stats.ttest_ind(k_med["Before1"], k_med["After1"])[1])
res["MannwhitneyuTest"] = pvals
res["TTest"] = ttest
```

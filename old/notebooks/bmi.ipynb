## BMI

```python
median_weights = pd.read_csv('/Volumes/GoogleDrive/My Drive/TAU/Code/DrugLab/data/mimiciii/1.4/preprocessed/median_weights.csv').drop(columns=['Unnamed: 0']).rename(columns={'VALUENUM':'WEIGHT'})
median_heights = pd.read_csv('/Volumes/GoogleDrive/My Drive/TAU/Code/DrugLab/data/mimiciii/1.4/preprocessed/median_heights.csv').drop(columns=['Unnamed: 0']).rename(columns={'HEIGHT_M':'HEIGHT'})
```

```python
median_weights
```

```python
median_heights
```

```python
merged = pd.merge(median_heights, median_weights, how='inner', on=['SUBJECT_ID', 'HADM_ID'])
```

```python
merged
```

```python
merged['CHARTTIME'] = merged.apply(lambda row: row['CHARTTIME_x'] if row['CHARTTIME_x']>row['CHARTTIME_y'] else row['CHARTTIME_y'], axis=1)
```

```python
merged['BMI'] = merged.apply(lambda row: row['WEIGHT']/(row['HEIGHT']**2) if (row['HEIGHT']!=0 and row['WEIGHT']!=0) else None, axis=1)
```

```python
merged
```

```python
merged = merged[merged['BMI']<184.69]
```

```python
merged.drop(columns=['CHARTTIME_x', 'CHARTTIME_y']).to_csv('/Volumes/GoogleDrive/My Drive/TAU/Code/DrugLab/data/mimiciii/1.4/preprocessed/bmi.csv')
```

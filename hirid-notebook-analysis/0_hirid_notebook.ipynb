## Before After Trend Calculation

```python
from scipy.stats import pearsonr, spearmanr
import numpy as np
def check_med2(row):
    if row["HADM_ID"] in t_med2["HADM_ID"].to_list():
        if row["ITEMID"] in t_med2[t_med2["HADM_ID"]==row["HADM_ID"]]["ITEMID"].to_list():
            return True
    return False

def get_med2(row):
    temp = t_med2[t_med2["HADM_ID"]==row["HADM_ID"]] 
    return temp[temp["ITEMID"]==row["ITEMID"]].iloc[0]
```

```python
from sklearn import datasets, linear_model, metrics

def get_normalized_trend(data):
    selected = data[['VALUENUM', 'hours_in']]
    if selected.shape[0]<2:
        return float("NaN")
    reg = linear_model.LinearRegression()
    reg.fit(np.array(data['hours_in']).reshape(-1,1), np.array(data['VALUENUM']).reshape(-1,1))
    return reg.coef_[0][0]
```

```python
window = (1,100, 5)
```

```python
def get_std_trends(r):

    row = r.copy()
    for i in range(window[0], window[1], window[2]):
        
        # before windows
        lab_vals = t_labs[t_labs["HADM_ID"]==row["HADM_ID"]]
        lab_vals = lab_vals[lab_vals["LabTimeFromAdmit"].dt.total_seconds()<row["MedTimeFromAdmit"].total_seconds()]

        b_window_start = row["MedTimeFromAdmit"].total_seconds() - (i*3600)
        b_window_end = row["MedTimeFromAdmit"].total_seconds() - (i+window[2])*3600
        lab_vals = lab_vals[lab_vals["LabTimeFromAdmit"].dt.total_seconds()<b_window_start]
        lab_vals = lab_vals[lab_vals["LabTimeFromAdmit"].dt.total_seconds()>b_window_end]

        t = lab_vals.groupby(["ITEMID"]).count()[["HADM_ID"]]
        val_counts_m = t[t["HADM_ID"]>=1]
        if val_counts_m.shape[0]==0:
            row[f"before_val_{i}_{window[2]}"] = {}
            row[f"before_mean_{i}_{window[2]}"] = {}
        else:
            l_m = lab_vals[lab_vals.ITEMID.isin(val_counts_m.index)]
            row[f"before_val_{i}_{window[2]}"] = l_m.sort_values("hours_in").groupby(["ITEMID"])[["VALUENUM", "hours_in"]].nth(-1)["VALUENUM"].dropna().to_dict()
            row[f"before_mean_{i}_{window[2]}"] = l_m.groupby(["ITEMID"])[["VALUENUM", "hours_in"]].mean()["VALUENUM"].dropna().to_dict()
        
        val_counts = t[t["HADM_ID"]>=2]
        if val_counts.shape[0]==0:
            row[f"before_trends_{i}_{window[2]}"] = {}
            row[f"before_std_{i}_{window[2]}"] = {}
        else:
            lab_vals = lab_vals[lab_vals.ITEMID.isin(val_counts.index)]  
            row[f"before_trends_{i}_{window[2]}"] = lab_vals[["ITEMID", "VALUENUM", "hours_in"]].dropna().groupby(["ITEMID"])[["VALUENUM", "hours_in"]].apply(lambda r : get_normalized_trend(r)).dropna().to_dict()
            row[f"before_std_{i}_{window[2]}"] = lab_vals.groupby(["ITEMID"])[["VALUENUM", "hours_in"]].std()["VALUENUM"].dropna().to_dict()
            
        # After windows
        lab_vals = labs[labs["HADM_ID"]==row["HADM_ID"]]
        med2_bool = check_med2(row)
        lab_vals = lab_vals[lab_vals["LabTimeFromAdmit"].dt.total_seconds()>row["MedTimeFromAdmit"].total_seconds()]
        a_window_start = row["MedTimeFromAdmit"].total_seconds() + (i*3600)
        a_window_end = row["MedTimeFromAdmit"].total_seconds() + (i+window[2])*3600
        lab_vals = lab_vals[lab_vals["LabTimeFromAdmit"].dt.total_seconds()>a_window_start]
        lab_vals = lab_vals[lab_vals["LabTimeFromAdmit"].dt.total_seconds()<a_window_end]
        if med2_bool:
            med2_val = get_med2(row)
            lab_vals = lab_vals[lab_vals["LabTimeFromAdmit"].dt.total_seconds()<med2_val["MedTimeFromAdmit"].total_seconds()]
        
        t = lab_vals.groupby(["ITEMID"]).count()[["HADM_ID"]]
        
        val_counts_m = t[t["HADM_ID"]>=1]
        if val_counts_m.shape[0]==0:
            row[f"after_val_{i}_{window[2]}"] = {}
            row[f"after_mean_{i}_{window[2]}"] = {}
        else:
            l_m = lab_vals[lab_vals.ITEMID.isin(val_counts_m.index)]
            row[f"after_val_{i}_{window[2]}"] = l_m.sort_values("hours_in").groupby(["ITEMID"])[["VALUENUM", "hours_in"]].nth(0)["VALUENUM"].dropna().to_dict()
            row[f"after_mean_{i}_{window[2]}"] = lab_vals.groupby(["ITEMID"])[["VALUENUM", "hours_in"]].mean()["VALUENUM"].dropna().to_dict()
        
        val_counts = t[t["HADM_ID"]>=2]
        if val_counts.shape[0]==0:
            row[f"after_trends_{i}_{window[2]}"] = {}
            row[f"after_std_{i}_{window[2]}"] = {}
        else:
            lab_vals = lab_vals[lab_vals.ITEMID.isin(val_counts.index)]   
            row[f"after_trends_{i}_{window[2]}"] = lab_vals[["ITEMID", "VALUENUM", "hours_in"]].dropna().groupby(["ITEMID"])[["VALUENUM", "hours_in"]].apply(lambda r : get_normalized_trend(r)).dropna().to_dict()
            row[f"after_std_{i}_{window[2]}"] = lab_vals.groupby(["ITEMID"])[["VALUENUM", "hours_in"]].std()["VALUENUM"].dropna().to_dict()
        
    return row
```

```python
def get_std_trends(r):

    row = r.copy()

    for i in range(1, 100, 5):
    
        lab_vals = labs[labs["HADM_ID"]==row["HADM_ID"]]
        lab_vals = lab_vals[lab_vals["LabTimeFromAdmit"].dt.total_seconds()<row["MedTimeFromAdmit"].total_seconds()]

        t = lab_vals.groupby(["ITEMID"]).count()[["HADM_ID"]]
        val_counts = t[t["HADM_ID"]>=2]
        if val_counts.shape[0]==0:
            row["before_val"] = {}
            row["before_trends"] = {}
            row["before_std"] = {}
        else:
            lab_vals = lab_vals[lab_vals.ITEMID.isin(val_counts.index)]        
            row["before_val"] = lab_vals.sort_values("hours_in").groupby(["ITEMID"])[["VALUENUM", "hours_in"]].nth(-1).dropna()["VALUENUM"].to_dict()
            row["before_trends"] = lab_vals[["ITEMID", "VALUENUM", "hours_in"]].dropna().groupby(["ITEMID"])[["VALUENUM", "hours_in"]].apply(lambda r : get_normalized_trend(r)).dropna().to_dict()
            row["before_std"] = lab_vals.groupby(["ITEMID"])[["VALUENUM", "hours_in"]].std()["VALUENUM"].dropna().to_dict()
            
        lab_vals = labs[labs["HADM_ID"]==row["HADM_ID"]]
        med2_bool = check_med2(row)
        lab_vals = lab_vals[lab_vals["LabTimeFromAdmit"].dt.total_seconds()>row["MedTimeFromAdmit"].total_seconds()]
        if med2_bool:
            med2_val = get_med2(row)
            lab_vals = lab_vals[lab_vals["LabTimeFromAdmit"].dt.total_seconds()<med2_val["MedTimeFromAdmit"].total_seconds()]
        t = lab_vals.groupby(["ITEMID"]).count()[["HADM_ID"]]
        val_counts = t[t["HADM_ID"]>=2]
        if val_counts.shape[0]==0:
            row["after_val"] = {}
            row["after_trends"] = {}
            row["after_std"] = {}
        else:
            lab_vals = lab_vals[lab_vals.ITEMID.isin(val_counts.index)]   
            row["after_val"] = lab_vals.sort_values("hours_in").groupby(["ITEMID"])[["VALUENUM", "hours_in"]].nth(0).dropna()["VALUENUM"].to_dict()
            row["after_trends"] = lab_vals[["ITEMID", "VALUENUM", "hours_in"]].dropna().groupby(["ITEMID"])[["VALUENUM", "hours_in"]].apply(lambda r : get_normalized_trend(r)).dropna().to_dict()
            row["after_std"] = lab_vals.groupby(["ITEMID"])[["VALUENUM", "hours_in"]].std()["VALUENUM"].dropna().to_dict()
        
    return row
```

### Pvals - Hard Threshold

```python
from scipy import stats
med_lab_all = final.copy()
pvals = []
ttest = []
counts = med_lab_all.groupby(["ITEMID", "MED_NAME", "LAB_ITEMID", "LAB_NAME"]).count()["HADM_ID"]
res_out = pd.DataFrame(counts)
# res_out = res_out[res_out["HADM_ID"]>100]
```

```python
for i in res_out.index:
    med = i[0]
    lab = i[2]
    k_med = med_lab_all[med_lab_all["ITEMID"]==med]
    k_med = k_med[k_med["LAB_ITEMID"]==lab]
    pvals.append(stats.mannwhitneyu(k_med["LabBeforeTrends"], k_med["LabAfterTrends"])[1])
    ttest.append(stats.ttest_ind(k_med["LabBeforeTrends"], k_med["LabAfterTrends"])[1])
res_out["MannwhitneyuTest"] = pvals
res_out["TTest"] = ttest
```

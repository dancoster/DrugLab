## Filter Labs

```python
labs = l_p_df[l_p_df["HADM_ID"].isin(med1["HADM_ID"])]
```

```python
labs = labs[labs['LabTimeFromAdmit'].dt.total_seconds()>0]
```

```python
labs = labs.sort_values("LabTimeFromAdmit")
```

```python
labs = labs[labs['VALUENUM'].notna()]
```

```python
labs
```

## Pairs

```python
def check_med2(row):
    if row["HADM_ID"] in med2["HADM_ID"]:
        if row["ITEMID"] in med2[med2["HADM_ID"]==row["HADM_ID"]]["ITEMID"]:
            return True
    return False

def get_med2(row):
    temp = med2[med2["HADM_ID"]==row["HADM_ID"]] 
    return temp[temp["ITEMID"]==row["ITEMID"]].iloc[0]

def get_lab_data(row, type):
    lab_vals = labs[labs["HADM_ID"]==row["HADM_ID"]]
    if type=="mean_before":
        lab_vals = lab_vals[lab_vals["LabTimeFromAdmit"].dt.total_seconds()<row["MedTimeFromAdmit"].total_seconds()]
        if lab_vals.shape[0]==0:
            return -1
        return lab_vals.groupby("HADM_ID").mean()["VALUENUM"].iloc[0]
    elif type=="mean_after":
        med2_bool = check_med2(row)
        lab_vals = lab_vals[lab_vals["LabTimeFromAdmit"].dt.total_seconds()>row["MedTimeFromAdmit"].total_seconds()]
        if med2_bool:
            med2_val = get_med2(row)
            lab_vals = lab_vals[lab_vals["LabTimeFromAdmit"].dt.total_seconds()<med2_val["MedTimeFromAdmit"].total_seconds()]
            if lab_vals.shape[0]==0:
                return -1
        return lab_vals.groupby("HADM_ID").mean()["VALUENUM"].iloc[0]
```

```python
t1 = med1.apply(lambda row : get_lab_data(row, "mean_before"), axis=1)
```

```python
med1["LabBefore"] = t1
```

```python
med1 = med1[med1["LabBefore"]!=-1]
```

```python
med1 = med1.drop(columns=["Unnamed: 0.1", "Unnamed: 0.2"])
```

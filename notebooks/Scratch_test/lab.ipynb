```python
import pandas as pd
import os
```

```python
admits = pd.read_csv(os.path.join(data, "mimiciii/1.4/raw", "ADMISSIONS.csv.gz"))
```

```python
labevents = pd.read_csv(os.path.join(data, "mimiciii/1.4/raw", "LABEVENTS.csv.gz"))
```

```python
d_labitems = pd.read_csv(os.path.join(data, "mimiciii/1.4/raw",'D_LABITEMS.csv.gz'))
```

```python
d_labitems
```

```python
labevents = pd.merge(labevents, d_labitems, how="inner", on="ITEMID")
```

```python
labevents
```

```python
labevents = pd.merge(labevents, admits, how="inner", on="HADM_ID")
```

```python
labevents
```

```python
def change_col_to_datetime(inputevents_mv, feature):
    inputevents_mv[feature] = pd.to_datetime(inputevents_mv[feature])
    return inputevents_mv
labevents = change_col_to_datetime(labevents, 'ADMITTIME')
labevents = change_col_to_datetime(labevents, 'CHARTTIME')
labevents["LabTimeFromAdmit"] = labevents['CHARTTIME']-labevents['ADMITTIME']
```
